```python
try:
    dict={"key1":"Udit","key2":"Kandpal"}
    print(dict["key3"])
except KeyError as e:
    print("KeyError is generated: ", e)
```

2. IndexError - It is generated when an attempt is made to access an item in a list at an index which is out of bounds.

```python
try:
    list=[1,2,3,4,5,]
    print(list[6])
except IndexError as e:
    print("ListIndex is out of bound: ", e)
```

##### Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is generated when a package/module is imported which is not originally present.

```python
try:
    import Udit
except ImportError as e:
    print("ImportError is generated: ", e)
```

```python
#Same ModuleNotFoundError is generated when a module is imported which is not installed or present actually. 
try:
    import Udit
except ModuleNotFoundError as e:
    print("ModuleNotFoundError is generated: ", e)
```
