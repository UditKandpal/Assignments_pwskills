#### Q. What is App routing in Flask? Why do we use app routes?

App routing is used to map a specific URL with the associated function that is intended to perform some task.  

This is done by defining the "app.route" decorator exposed by the application instance.  

A route is the association between a URL and the function that handles it.

#### Q. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for building URLs dynamically. The first argument of the url_for() function is the name of the specified function, and then you can pass any number of keyword arguments corresponding to the variable part of the URL.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/profile/<username>")
def profile(username):
    if username=='udit':
        return redirect(url_for('profile', username ='udit'))

if __name__=="__main__":
    app.run(host="0.0.0.0", port=3000)
```
